### Las columnas muestran la información en diccionarios. Se necesita extraer el valor correcto de esos diccionarios.

```python
if response.status_code == 200:
    # Convertir la respuesta en JSON
    datos = response.json()
    
    # El segundo elemento de la lista contiene los datos
    observaciones = datos[1]
    
    # Se crea un DataFrame(observaciones)
    datos = pd.DataFrame(observaciones)
    
    # Extraer el nombre del país del diccionario 'country'
    datos['Pais'] = datos['country'].apply(lambda x: x['value'])
    
    # Seleccionar las columnas que necesitas
    datos = datos[['Pais', 'date', 'value']]
    datos.columns = ['Pais', 'Año', 'INB_dólares']
    
    # Se convierten las variables a tipo numérico
    datos['Año'] = pd.to_numeric(datos['Año'])
    datos['INB_dólares'] = pd.to_numeric(datos['INB_dólares'])
    
    # Eliminar filas sin datos
    datos = datos.dropna()
    
    # Crear columna en miles de millones para mejor visualización
    datos['INB_miles_millones'] = datos['INB_dólares'] / 1000000000
    
    print(f"\nDatos obtenidos: {len(datos)} registros")
    print(f"\nRango de años: {datos['Año'].min()} - {datos['Año'].max()}")
    print(f"\nPaíses: {list(datos['Pais'].unique())}")
    
    # Mostrar primeros registros con formato legible
    print("\nMuestra de datos (en miles de millones de dólares):")
    print(datos[['Pais', 'Año', 'INB_miles_millones']].head(10))
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation
import seaborn as sns
from IPython.display import HTML

# Configurar estilo
sns.set_style("whitegrid")

# Filtrar solo Centroamérica
paises_centroamerica = ['Costa Rica', 'El Salvador', 'Guatemala', 'Honduras', 'Nicaragua']
datos_filtrados = datos[datos['Pais'].isin(paises_centroamerica)]

# Pivotear los datos
datos_pivot = datos_filtrados.pivot(index='Año', columns='Pais', values='INB_dólares')

# Colores
colores = {
    'Costa Rica': '#2E86AB',
    'Guatemala': '#F18F01',
    'Honduras': '#C73E1D',
    'Nicaragua': '#6A994E',
    'El Salvador': '#6D597A'
}

# Crear la figura y el eje
fig, ax = plt.subplots(figsize=(14, 8))

# Obtener años y países
años = datos_pivot.index.tolist()
paises = datos_pivot.columns.tolist()

# Configurar límites de los ejes
ax.set_xlim(2000, 2025)
ax.set_ylim(0, datos_pivot.max().max() / 1_000_000_000 * 1.1)

# Configurar el eje Y
def formato_miles_millones(x, p):
    return f'${x:.0f} mil M'

ax.yaxis.set_major_formatter(plt.FuncFormatter(formato_miles_millones))

# Títulos
ax.set_title('Comportamiento del INB en Centroamérica (2000-2024)', 
             fontsize=18, fontweight='bold', pad=20)
ax.set_xlabel('Año', fontsize=13, fontweight='bold')
ax.set_ylabel('INB (miles de millones de US$)', fontsize=13, fontweight='bold')
ax.grid(True, linestyle='--', alpha=0.3)

# Inicializar líneas vacías
lines = {}
for pais in paises:
    line, = ax.plot([], [], 
                    label=pais,
                    color=colores[pais],
                    linewidth=3.5 if pais == 'Honduras' else 2,
                    alpha=1.0 if pais == 'Honduras' else 0.7)
    lines[pais] = line

# Añadir leyenda
ax.legend(loc='upper left', fontsize=11)

# Texto para mostrar el año actual
year_text = ax.text(2022, datos_pivot.max().max() / 1_000_000_000 * 0.9, 
                    '', fontsize=14, fontweight='bold', ha='center')

# Agregar fuente y créditos (se mostrarán estáticos)
fuente_text = ax.text(0.02, -0.12, 
                      'Fuente: Banco Mundial (2024) | Elaboración propia', 
                      transform=ax.transAxes, fontsize=10, 
                      color='#7F8C8D', ha='left')

# Función de inicialización
def init():
    for pais in paises:
        lines[pais].set_data([], [])
    year_text.set_text('')
    return list(lines.values()) + [year_text, fuente_text]

# Función de actualización (se ejecuta para cada año)
def update(frame):
    año_actual = años[frame]
    
    for pais in paises:
        # Datos desde 2000 hasta el año actual
        años_hasta = [a for a in años if a <= año_actual]
        valores_hasta = [datos_pivot.loc[a, pais] / 1_000_000_000 for a in años_hasta]
        lines[pais].set_data(años_hasta, valores_hasta)
    
    # Actualizar texto del año
    year_text.set_text(f'Año: {año_actual}')
    
    return list(lines.values()) + [year_text, fuente_text]

# Crear la animación
ani = FuncAnimation(fig, update, frames=range(len(años)), 
                    init_func=init, blit=True, interval=300, repeat=True)

# Mostrar la animación usando to_jshtml() (no requiere ffmpeg)
plt.close()  # Evita que se muestre el gráfico estático
HTML(ani.to_jshtml())
```
